### 8. Missing Values
#### Berapa banyak NaN age data dan mean dari setiap who?
#### Isi data NaN dari kolom age dengan mean dari umur tergantung whonya dia apa (child/woman/man)
#### Kemudian print banyak NaN age data dan mean umur setiap who lagi.

```python
nanMan = len(data[(data['who']=='man') & (data['age'].isnull())])
meanMan = data[(data['who']=='man')]['age'].mean()

nanWoman = len(data[(data['who']=='woman') & (data['age'].isnull())])
meanWoman = data[(data['who']=='woman')]['age'].mean()

nanChild = len(data[(data['who']=='child') & (data['age'].isnull())])
meanChild = data[(data['who']=='child')]['age'].mean()

print('Banyak Nan man & mean: '+ str(nanMan) + ' & ' + str(meanMan))
print('Banyak Nan woman & mean: '+ str(nanWoman) + ' & ' + str(meanWoman))
print('Banyak Nan child & mean: '+ str(nanChild) + ' & ' + str(meanChild))
```

```python
def applyMean(x):
    if pd.notnull(x['age']):
        return x['age']
    else:
        if (x['who']=='man'):
            return round(meanMan)
        elif (x['who']=='woman'):
            return round(meanWoman)
        elif (x['who']=='child'):
            return round(meanChild)
```

```python
data['age'] = data.apply(applyMean,axis=1)
```

```python
nanMan = len(data[(data['who']=='man') & (data['age'].isnull())])
meanMan = data[(data['who']=='man')]['age'].mean()

nanWoman = len(data[(data['who']=='woman') & (data['age'].isnull())])
meanWoman = data[(data['who']=='woman')]['age'].mean()

nanChild = len(data[(data['who']=='child') & (data['age'].isnull())])
meanChild = data[(data['who']=='child')]['age'].mean()

print('Banyak Nan man & mean: '+ str(nanMan) + ' & ' + str(meanMan))
print('Banyak Nan woman & mean: '+ str(nanWoman) + ' & ' + str(meanWoman))
print('Banyak Nan child & mean: '+ str(nanChild) + ' & ' + str(meanChild))
```

#### Isi NaN embarked dan embark_town dan deck dengan modus masing2 (print modus masing2 juga)

```python
modEmbarked = data['embarked'].mode().values[0]
print('Modus embarked = ' + str(modEmbarked) )
```

```python
def applyEmbarked(x):
    if pd.notnull(x['embarked']):
        return x['embarked']
    else:
        return modEmbarked

def applyEmbarkTown(x):
    if pd.notnull(x['embark_town']):
        return x['embark_town']
    else:
        return modEmbarkedTown

def applyDeck(x):
    if pd.notnull(x['deck']):
        return x['deck']
    else:
        return modDeck
```

```python
data['embarked'] = data.apply(applyEmbarked,axis=1)
```

```python
data['embark_town'] = data.apply(applyEmbarkTown,axis=1)
```

```python
data['deck'] = data.apply(applyDeck,axis=1)
```

### 9. Adding New Features/Columns/Variables

#### Buat DataFrame baru dengan column2 sebagai berikut
 - __id__ = id table ini (PK)
 - __mean_fare__ = Mean Fare
 - __std_fare__ = Std Fare
 - __batas_min_fare__ = Mean Fare - (3 x Std Fare)
 - __batas_max_fare__ = Mean Fare + (3 x Std Fare)
 
#### Dan data2nya ada 3 (fare setiap class yaitu first, second, and third class)

```python
print("['First' 'Second' 'Third']")

id = [1,2,3]
mean_fare = []
std_fare = []
batas_min_fare = []
batas_max_fare = []
for item in id:
    mean_fare.append(data[data['pclass']==item]['fare'].mean())
    std_fare.append(data[data['pclass']==item]['fare'].std())
    batas_min_fare.append((data[data['pclass']==item]['fare'].mean())-(3*(data[data['pclass']==item]['fare'].std())))
    batas_max_fare.append((data[data['pclass']==item]['fare'].mean())+(3*(data[data['pclass']==item]['fare'].std())))

newdict = {
    'id': id,
    'mean_fare': mean_fare,
    'std_fare': std_fare,
    'batas_min_fare': batas_min_fare,
    'batas_max_fare': batas_max_fare
}

pd.DataFrame(newdict)
```

```python
batas_min_fare[0]
```

#### Print outlier % of each class

```python
classCat = ['First', 'Second','Third']
for item in range (1,4):
    out = ('Outlier % '+ str(classCat[item-1])+ ' : ' + str(len(data[(data['pclass']==item) & ((data['fare']>batas_max_fare[item-1]) | (data['fare']<batas_min_fare[item-1]))])/len(data[data['pclass']==item])*100))
    print(out)
```

#### Tambahkan column outlier (1 for yes and 0 for no)

```python
def cekOutlier(x):
    if x['pclass']==1:
        if ((x['fare']>batas_max_fare[0]) | (x['fare']<batas_min_fare[0])):
            return 1
        else:
            return 0
    elif x['pclass']==2:
        if ((x['fare']>batas_max_fare[1]) | (x['fare']<batas_min_fare[1])):
            return 1
        else:
            return 0
    elif x['pclass']==3:
        if ((x['fare']>batas_max_fare[2]) | (x['fare']<batas_min_fare[2])):
            return 1
        else:
            return 0
```

```python
data['outlier'] = data.apply(cekOutlier,axis=1)
```
